Preparing inputs and outputs

```python
X= iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
y = iris[['Species']].values
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

Accuracy

```python
model.score(X,y)
```

Make Predication

```python
expected = y
predicated = model.predict(X)
predicated
```

Summarize the fit of the model

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(expected,predicated))
```

```python
print(metrics.confusion_matrix(expected,predicated))
```

```python
#define a function that counts the number of times each genre appear:
def count_word(df, ref_col, liste):
    keyword_count = dict()
    for s in liste: keyword_count[s] = 0
    for liste_keywords in df[ref_col].str.split('|'):
        if type(liste_keywords) == float and pd.isnull(liste_keywords): continue
        for s in liste_keywords:
            if pd.notnull(s): keyword_count[s] += 1
    # convert the dictionary in a list to sort the keywords  by frequency
    keyword_occurences = []
    for k,v in keyword_count.items():
        keyword_occurences.append([k,v])
    keyword_occurences.sort(key = lambda x:x[1], reverse = True)
    return keyword_occurences, keyword_count
```

```python
#here we  make census of the genres:
genre_labels = set()
for s in data['genres'].str.split('|').values:
    genre_labels = genre_labels.union(set(s))
```

```python
#counting how many times each of genres occur:
keyword_occurences, dum = count_word(data, 'genres', genre_labels)
keyword_occurences
```

 I've explain the concept of training data and test data using a simple analogy:


---



**Training Data** is like a student practicing math problems from a textbook before an exam. The student learns from the textbook, solves various math problems, and understands different problem-solving techniques. This is the data that the student "trains" on, just like a machine learning model "learns" from training data.

**Test Data** is like the actual exam. During the exam, the student faces new, unseen math questions that are not from the textbook. The student uses the knowledge gained from practicing on the training data (the textbook) to solve these new questions and predict the answers.

In the context of machine learning:
- **Training Data** is the part of the dataset used to teach the machine learning model. It's like the textbook that the model learns from.
- **Test Data** is the part of the dataset that the model hasn't seen during training. It's like the new exam questions that the model needs to answer based on what it learned from the training data.

The goal of training and testing is to ensure that the machine learning model can perform well not just on the problems it has already seen (training data, like textbook exercises) but also on new, unseen problems (test data, like exam questions). This helps evaluate how well the model can generalize its learning to real-world situations.

    2. training_data_accuracy = accuracy_score(X_train_predication, Y_train) "

    The result, ' training_data_accuracy ',
    represents the percentage of correct predictions made by the model on the training data. It's a measure of how well the model has learned from the data it was trained on.
